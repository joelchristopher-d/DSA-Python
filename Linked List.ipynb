```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        
    def push(self,data):
        newnode = Node(data)
        newnode.next = self.head
        self.head = newnode
        
    def Append(self,data):
        newnode = Node(data)
        if self.head is None:
            self.head = newnode
            return
        last = self.head
        while(last.next):
            last = last.next
        last.next = newnode
        
    def insertAt(self,prev,data):
        if prev is None:
            print("prev is none")
        newnode = Node(data)
        newnode.next = prev.next
        prev.next = newnode
        
    def PrintList(self):
        last = self.head
        while(last):
            print(last.data,end=" ")
            last = last.next
            
    def insert(self,key,data):
        newnode = Node(data)
        curr = self.head
        while(curr is not None):
            if curr.data == key:
                break
#             prev = curr
            curr = curr.next
        newnode.next = curr.next
        curr.next = newnode
            
        
    def Delete(self,key):
        temp = self.head
        if temp is not None:
            if temp.data == key:
                self.head = temp.next
                temp = None
            
        while(temp is not None):
            if(temp.data == key):
                break
            prev = temp
            temp = temp.next
            
        if (temp == None):
            return
        prev.next = temp.next
        temp = None
        
    def GetCountNode(self,node):
        if(not node):
            return 0
        else:
            return 1 + self.GetCountNode(node.next)
        
    def count(self):
        return self.GetCountNode(self.head)
            
    def Reverse(self):
        prev = None
        nxt = None
        curr = self.head
        while(curr):
            nxt = curr.next
            curr.next = prev
            
            prev = curr
            curr = nxt
        self.head = prev
            
if __name__ == "__main__":
    llist = LinkedList()
    llist.push(3)
    llist.push(1)
    llist.Append(9)
    llist.push(8)
    llist.insertAt(llist.head,7)
    llist.insert(7,2)
    llist.Delete(9)
    llist.PrintList()
#     print("\ncount:",llist.GetCountNode())
    print("\ncount:",llist.count())
    
    llist.Reverse()
    llist.PrintList()
```
